```python
data1={'key1':[1,45,67,89,578],
       'key2':[123,456,778,34,99],
       'key3':[11,22,33,44,55]
}
```

```python
data2={'key1':[1,145,607,809,58],
       'key4':[12,56,78,3,9],
       'key5':[111,252,373,447,505]
}
```

```python
data1=pd.DataFrame(data1)
```

```python
data2=pd.DataFrame(data2)
```

```python
pd.merge(data1,data2)#by default how=inner
```

```python
pd.merge(data1,data2,how='left')#it returns the proper left
```

```python
pd.merge(data1,data2,how='right')# it returns the proper right 
```
